```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
df = pd.read_csv("house_prices_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
X = df.area.values.reshape(-1,1)
y = df.price.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Training
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict
y_pred = model.predict(X_test)
```

```python
# Evaluation of created model
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R^2 Score: {r2}")
```

```python
# When residual is plotted randomly , model is good 

residuals = y_test - y_pred

plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted Prices')
plt.ylabel('Residuals')
plt.title('Residual Analysis')
plt.show()
```

```python
# Plotting prediction

plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.plot(X_test, y_pred, color='red', label='Model Prediction')
plt.xlabel('Size of the house (sq ft)')
plt.ylabel('Price of the house')
plt.title('Training Data and Model Prediction')
plt.legend()
plt.show()
```
